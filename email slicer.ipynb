```python
class email_slice:
    def __init__(self):
        email = input("What is your email address?: ").strip()
        self.email = email
    def show(self):
        user_name = self.email[:self.email.index("@")]
        domain_name = self.email[self.email.index("@")+1:]
        print("Your username is : ",user_name)
        print("Your password is :",domain_name)
```

```python
test = email_slice()
test.show()
```
